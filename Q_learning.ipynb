```python
import pickle
from collections import defaultdict
import gym
import numpy as np
import time
```

```python
# Q-table
# state | action 0 | action 1 | action 2
Q = defaultdict(lambda: [0,0,0])
```

```python
env = gym.make('MountainCar-v0')

# 把浮点数状态转换为整数 [0,50]
def transform_state(state):
    pos, v = state
    pos_low, v_low = env.observation_space.low
    pos_high, v_high = env.observation_space.high

    a = 50 * (pos-pos_low) / (pos_high-pos_low)
    b = 50 * (v-v_low) / (v_high-v_low)

    return int(a), int(b)
```

```python
# Q[s][a] = (1-lr) * Q[s][a] + lr * (reward + factor * max(Q[next_s]))

lr, factor = 0.7, 0.95
episodes = 10000
score_list = []

for i in range(episodes):
    state, _ = env.reset()
    s = transform_state(state)
    score = 0

    while True:
        a = np.argmax(Q[s])
        # 在训练刚开始的时候，增加随机性
        if np.random.random() > i *3 / episodes:
            a = np.random.choice([0,1,2])

        # 执行动作
        next_s, reward, done, truncated, _ = env.step(a)
        next_s = transform_state(next_s)

        # 更新Q-Table
        Q[s][a] = (1-lr) * Q[s][a] + lr * (reward + factor * max(Q[next_s]))
        score += reward
        s = next_s

        if done or truncated:
            score_list.append(score)
            print(f"episode: {i+1}, score: {score}, max score: {max(score_list)}")
            break


env.close()

# 保存模型
with open('Mountain_car_q_learning.pickle','wb') as f:
    pickle.dump(dict(Q), f)
    print("model saved!")
```
